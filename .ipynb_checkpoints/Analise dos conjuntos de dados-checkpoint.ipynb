```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from string import punctuation
import plotly.graph_objects as go
import nltk
```

```python
# Leitura dos csv
df_course = pd.read_csv('data/course.csv')
df_offers = pd.read_csv('data/offers.csv')
```

```python
# Drop da coluna ID e ajustes das outras colunas
df_course = df_course.drop("“ID”", axis=1)
df_course.columns = ['DISCIPLINA','CONTEUDO']
df_offers = df_offers.drop("“ID”", axis=1)
df_offers.columns = ['TITLE','VALORDESCRIPTION']
```

## Analise os dados dos dois conjuntos 

```python
# Criacao de stopwords e pontuacao
stop_words = set(stopwords.words('portuguese') + list(punctuation))
```

```python
 # Função para criar word cloud e rank de palavras
def generateWC(text):
    wordcloud = WordCloud(stopwords=stop_words,max_font_size=50, max_words=100, background_color="white").generate(text)
    plt.figure(figsize=(12,8))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    
def getWordRank(text):
    wordcloud = WordCloud(stopwords=stop_words,max_font_size=50, max_words=100, background_color="white")
    rank = wordcloud.process_text(text)
    sort_rank = sorted(rank.items(), key=lambda x: x[1], reverse=True) 
    return sort_rank

def generateTop10Plot(text):
    top10 = getWordRank(text)[:11]
    text = list(zip(*top10))[0]
    value = list(zip(*top10))[1]
    fig = go.Figure(data=[go.Bar(
        x=text,
        y=value
    )])
    fig.update_layout(title_text='Top 10 ')
    fig.show()
```

```python
# Add string vazia nos valores NA da colubna conteudo
df_course['CONTEUDO'].fillna('', inplace=True)
```

#### Wordcloud e rank geral

```python
# Wordcloud geral da coluna CONTEUDO 
text_conteudo = " ".join(str(desc.lower()) for desc in df_course['CONTEUDO'])
generateWC(text_conteudo) 
```

```python
# Rank geral dos assuntos mais frequentes da coluna CONTEUDO 
getWordRank(text_conteudo)
```

```python
# Top 10 dos assuntos mais frequentes
generateTop10Plot(text_conteudo)
```

#### Wordcloud e rank por disciplina

```python
# Rank da coluna CONTEUDO por disciplina
disciplinas = np.unique(df_course['DISCIPLINA'])
for disc in disciplinas:
    df_disciplina = df_course[df_course['DISCIPLINA'] == disc]
    text_disc = " ".join(str(desc.lower()) for desc in df_disciplina['CONTEUDO'])
    print(disc)
    generateWC(text_disc)
    generateTop10Plot(text_disc)
    print('-----------------------------------------------------------------------------------------------------')
```
